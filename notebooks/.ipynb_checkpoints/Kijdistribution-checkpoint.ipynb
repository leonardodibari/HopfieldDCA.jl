```julia
function att_corr(K)
    H = size(K,3)
    pears = zeros(H,H)
    for h in 1:H
        for y in h:H
            pears[h,y] = cor(K[:,:,h][:],K[:,:,y][:])
        end
    end
    matshow(pears)
    colorbar()
end
```
